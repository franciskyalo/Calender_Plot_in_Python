```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl="en-US", tz=360)
pytrends.build_payload(kw_list=["data science"])

df = pytrends.interest_over_time()
```

```python
df.head()
```

```python
data_to_use = df['data science']
```

```python
data_to_use.head()
```

```python
!pip install calplot
```

```python
import calplot
import matplotlib.pyplot as plt

calplot.calplot(data_to_use, yearascending = False,
                suptitle = 'Calendar with years in descending order',
                suptitle_kws = {'x': 0.5, 'y': 1.0})
# plt.show()
```

```python
# prompt: extra the day from the data_to_use dataframe above, e.g if it is sunday or monday in terms of days of the week

import pandas as pd

# Assuming 'data_to_use' is a pandas Series as in the previous code
# If it's a DataFrame, replace 'data_to_use' with 'data_to_use['data science']'

# Convert the index to datetime objects if it's not already
data_to_use.index = pd.to_datetime(data_to_use.index)

# Extract the day of the week
data_to_use['day_of_week'] = data_to_use.index.day_name()

# Display the first few rows to show the new column
print(data_to_use.head())
```

```python
import numpy as np; np.random.seed(1)
import pandas as pd

# Sequence of dates
days = pd.date_range('01/01/2020', periods = 1277, freq = 'D')

# Pandas time series with random data for each day
values = pd.Series(np.random.randn(len(days)), index = days)
```

```python

calplot.calplot(values,cmap='YlGn',colorbar=False)
# plt.show()
```
